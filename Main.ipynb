```python
from sklearn.impute import SimpleImputer
columns_to_impute = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
imputer = SimpleImputer(strategy='mean')
df[columns_to_impute] = imputer.fit_transform(df[columns_to_impute])
df
```

```python
df['profit']=df['Close']-df['Open']
df
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
x= df[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 'year']]
y= df['profit'].to_list()
scaler = StandardScaler()
x = scaler.fit_transform(x)
svr = SVR(kernel='linear', C=1.0)
svr.fit(x,y)
y_pred = svr.predict(x)
```

```python
s = [32, 35,10,56,554,40.33, 2024] 
s = scaler.transform([s])
predicted_profit = svr.predict(s)
print(f"Predicted Profit: {predicted_profit[0]}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f"Mean Squared Error: {mse:}")
print(f"R-squared (Coefficient of Determination): {r2:.2f}")
```
